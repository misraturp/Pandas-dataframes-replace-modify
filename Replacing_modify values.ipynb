```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("NYC_Jobs.csv")
data.head()
```

```python
data.fillna('kitten')
```

```python
data.replace("NYC HOUSING AUTHORITY","NYC AUTHORITY FOR HOUSING")
data.replace("NYC HOUSING AUTHORITY",'NYC_Hous_A', inplace=True)
```

```python
# Changes the value in the original dataframe

data.loc[data["Posting Type"] == "External", "Posting Type"] = 1
data.at[data["Posting Type"] == 1,'Civil Service Title']='CITA'
data.loc[0, "Posting Type"] = "Unknown" #works with at too
```

```python
data.loc[0, "Posting Type"] = 12
```

```python
data
```

```python
data["is_LegalAffairs"] = np.where(data["Job Category"] == "Legal Affairs", 1, 0)
```

```python
data
```

```python
data['Civil Service Title'].mask(data['Civil Service Title'] == 'SENIOR IT ARCHITECT', 0, inplace=False)
```

```python
data
```

```python
def capitalize(text):
    return text.lower().capitalize()

data['Civil Service Title'].apply(capitalize)
```

```python
def replace_values(text):
    if text == 'NYC EMPLOYEES RETIREMENT SYS':
        return 'Retirement SYS'
    elif text == 'HUMAN RIGHTS COMMISSION':
        return 'HRC'
    else:
        return text

data['Agency'] = data['Agency'].apply(replace_values)
```

```python
data
```

```python
data['Agency'].map({'NYC HOUSING AUTHORITY':'NYC_H_A'})
```

```python
def add_year(value):
    return str(value) + '_2022'
    

data[['Job ID', 'Civil Service Title']] = data[['Job ID', 'Civil Service Title']].applymap(add_year)
```

```python
data
```

```python
data['Salary range'] = pd.cut(data['Salary Range From'], 3)
data
```
